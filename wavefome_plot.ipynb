## Imports

```python
from obspy import read
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Load data

```python
st = read('high_res_for_Chen/*.sac')
delta_t=st[0].stats.sac['delta']
```

```python
st
```

```python
Amax=max(abs(st[0][:]))
```

## Plot waveforms

```python
size=32
params = {'legend.fontsize': size*0.75,
          'figure.figsize': (16,12),
          'axes.labelsize': size,
          'axes.titlesize': size,
          'xtick.labelsize': size*0.75,
          'ytick.labelsize': size*0.75,
          'axes.titlepad': 25}
#color_cycle = plt.rcParams['axes.prop_cycle'].by_key()['color']
fig=plt.rcParams.update(params)
plt.plot(np.arange(0,delta_t*st[0].stats.npts,delta_t),st[0][:]/Amax, color='C0',lw=3,label='Layered')
plt.plot(np.arange(0,delta_t*st[1].stats.npts,delta_t),st[1][:]/Amax, color='C1',lw=3,label='3D')
plt.xlabel('Time (sec)')
plt.ylabel('Normalized amplitudes (A.U.)')
plt.xlim([0,3])
plt.grid(True)
plt.legend(loc="upper right")
plt.savefig('waveforms_layered_3d_vel.pdf',dpi=300)
plt.savefig('waveforms_layered_3d_vel.eps',dpi=300)
```
